## Prep

```python
import os
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision
from torchvision import transforms
from PIL import Image
import rasterio

# Define the mapping from class codes to label indices
class_mapping = {
    111: 0,
    112: 1,
    121: 2,
    122: 3,
    123: 4,
    124: 5,
    131: 6,
    132: 7,
    133: 8,
    141: 9,
    142: 10,
    211: 11,
    212: 12,
    213: 13,
    221: 14,
    222: 15,
    223: 16,
    231: 17,
    241: 18,
    242: 19,
    243: 20,
    244: 21,
    311: 22,
    312: 23,
    313: 24,
    321: 25,
    322: 26,
    323: 27,
    324: 28,
    331: 29,
    332: 30,
    333: 31,
    334: 32,
    335: 33,
    411: 34,
    412: 35,
    421: 36,
    422: 37,
    423: 38,
    511: 39,
    512: 40,
    521: 41,
    522: 42,
    523: 43
}

num_classes = len(class_mapping)

class SatelliteDataset(Dataset):
    def __init__(self, image_dir, reference_map_base_path, transform=None):
        self.image_dir = image_dir
        self.reference_map_base_path = reference_map_base_path
        self.image_files = [f for f in os.listdir(image_dir) if f.endswith('.png')]
        self.transform = transform

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        image_file = self.image_files[idx]
        patch_id = os.path.splitext(image_file)[0]

        # Load the image
        image_path = os.path.join(self.image_dir, image_file)
        image = Image.open(image_path).convert("RGB")

        # Load the reference map
        parent_folder = "_".join(patch_id.split("_")[:-2])
        reference_map_path = os.path.join(self.reference_map_base_path, parent_folder, patch_id, f"{patch_id}_reference_map.tif")
        
        with rasterio.open(reference_map_path) as src:
            reference_map = src.read(1)  # Reading the first band

        # Create a multi-label vector from the reference map
        unique_classes = np.unique(reference_map)
        multi_label_vector = np.zeros(num_classes)
        for class_code in unique_classes:
            if class_code in class_mapping:
                multi_label_vector[class_mapping[class_code]] = 1

        if self.transform:
            image = self.transform(image)
        
        return image, torch.tensor(multi_label_vector, dtype=torch.float32), patch_id

# Data transforms (if needed)
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

from torch.utils.data import DataLoader, random_split

# Define your dataset
image_dir = r"C:\Users\CL-11\OneDrive\Research\SateliteSR\Final\SampleDataset"
reference_map_base_path = r"C:\Datasets\BigEarthNet-S2\Reference_Maps\Reference_Maps"
dataset = SatelliteDataset(image_dir, reference_map_base_path, transform=transform)

# Calculate dataset sizes
train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size  # Ensure the remaining data goes to the test set

# Split the dataset into training, validation, and test sets
train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])

# Create DataLoader instances for each set
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print(len(train_loader), len(val_loader), len(test_loader))
```

```python
import matplotlib.pyplot as plt
import torchvision
import numpy as np

# Unnormalization function (reversing the normalization applied earlier)
def unnormalize(img, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]):
    img = img.clone()  # Clone the tensor to avoid modifying the original
    for t, m, s in zip(img, mean, std):
        t.mul_(s).add_(m)  # Denormalize: multiply by std and add mean
    return img

# Function to display the image
def imshow(img):
    img = unnormalize(img)     # Unnormalize the image
    npimg = img.numpy()        # Convert to NumPy array
    plt.imshow(np.transpose(npimg, (1, 2, 0)))  # Transpose to (H, W, C) format
    plt.show()

# Sample a batch from the dataloader
sample = next(iter(train_loader))

images, labels, id = sample  # Get images and labels from the batch
imshow(torchvision.utils.make_grid(images))  # Show a grid of images
```

```python
import torch
import torch.nn as nn
from torchvision import models

# Load the pre-trained ResNet-50 model
model = models.resnet50(pretrained=True)

# Modify the final layer to match the number of classes
model.fc = nn.Linear(model.fc.in_features, num_classes)

# Move model to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
```

```python
criterion = nn.BCEWithLogitsLoss()  # Multi-label classification loss
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
from tqdm import tqdm 

num_epochs = 10  # You can adjust this based on your dataset size
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for images, labels, id in tqdm(train_loader):
        images, labels = images.to(device), labels.to(device)

        # Zero the gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(images)

        # Compute loss
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}")
```

## Eval

```python
import torch
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model, dataloader):
    model.eval()  # Set model to evaluation mode
    all_preds = []
    all_labels = []

    with torch.no_grad():  # Disable gradient tracking
        for images, labels, id in dataloader:
            images, labels = images.to(device), labels.to(device)

            # Forward pass
            outputs = model(images)

            # Apply sigmoid to get probabilities
            preds = torch.sigmoid(outputs) > 0.5  # Threshold at 0.5 for multi-label classification

            # Append predictions and labels to lists
            all_preds.append(preds.cpu().numpy())
            all_labels.append(labels.cpu().numpy())

    all_preds = np.vstack(all_preds)
    all_labels = np.vstack(all_labels)

    return all_preds, all_labels

# Calculate evaluation metrics
def calculate_metrics(preds, labels):
    accuracy = accuracy_score(labels, preds)
    precision = precision_score(labels, preds, average='macro')  # Use 'macro' for multi-label
    recall = recall_score(labels, preds, average='macro')
    f1 = f1_score(labels, preds, average='macro')

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")


# Assume test_dataloader is your test dataset dataloader
preds, labels = evaluate_model(model, test_loader)

# Calculate metrics
calculate_metrics(preds, labels)
```

```python
import torch
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def compute_confusion_matrices(model, dataloader, num_classes):
    # Set the model to evaluation mode
    model.eval()

    # Initialize lists to hold all predictions and true labels
    all_preds = []
    all_labels = []

    # Disable gradient computation for evaluation
    with torch.no_grad():
        for images, labels, ids in dataloader:  # Go through the full test_loader
            images, labels = images.to(device), labels.to(device)
            
            # Forward pass to get predictions
            outputs = model(images)
            
            # Apply sigmoid and threshold at 0.5 to get predicted labels
            preds = (torch.sigmoid(outputs) > 0.5).int()

            # Append predictions and labels (move to CPU)
            all_preds.append(preds.cpu().numpy())
            all_labels.append(labels.cpu().numpy())
    
    # Concatenate the results to create full prediction and label arrays
    all_preds = np.vstack(all_preds)  # Convert list of batches into a full array
    all_labels = np.vstack(all_labels)

    # Compute confusion matrices for each class
    confusion_matrices = []
    for i in range(num_classes):
        cm = confusion_matrix(all_labels[:, i], all_preds[:, i])
        confusion_matrices.append(cm)

    return confusion_matrices

# Use the test_loader and the number of classes to compute confusion matrices
num_classes = test_loader.dataset[0][1].shape[0]  # Assuming second element is labels, e.g., shape [4] for 4 classes
print(num_classes)
confusion_matrices = compute_confusion_matrices(model, test_loader, num_classes)

def plot_confusion_matrix(cm, class_name):
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, square=True)
    plt.title(f'Confusion Matrix for {class_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

# Plot confusion matrices for each class
class_names = list(class_mapping.values())
for i, cm in enumerate(confusion_matrices):
    plot_confusion_matrix(cm, class_names[i])
```
